```python
from google_play_scraper import app, Sort, reviews
import pandas as pd
import os
os.chdir(r"C:\Users\mohds\Downloads\Sem -3\NLP Project")
```

```python
def scrape_google_playstore_reviews(package_name, count = 2000, lang = 'en', country = 'IN', sort = Sort.NEWEST):

    # scraping review from Play_Store
    g_reviews = reviews(
            package_name,
            count,
            lang, # defaults to 'en'
            country, # defaults to 'us'
            sort, # defaults to Sort.NEWEST
        )

    # Creating DataFrame from reviews data
    reviews_data = g_reviews[0]
    df = pd.DataFrame(reviews_data)

    df.drop(columns={'userImage', 'reviewId','reviewCreatedVersion'},inplace = True)
    df.rename(columns= {'score': 'rating', 
                        'userName': 'user_name', 
                        'content': 'review_description', 
                        'at': 'review_date', 
                        'replyContent': 'developer_response', 
                        'repliedAt': 'developer_response_date', 
                        'thumbsUpCount': 'thumbs_up'},inplace = True)
    df.insert(loc=0, column='source', value='Google Play')
    df.insert(loc=1, column='Appliction', value=package_name)

    # Dropping duplicate reviews
    df.drop_duplicates(['review_description'], inplace=True)
    df.reset_index(inplace=True)
    df.drop('index', axis = 1, inplace=True)

    return df

def export_to_csv(file_name, data_frame):
    data_frame.to_csv(file_name, index = False)
```

```python
package_name = input("Enter the package name (e.g: com.airbnb.android): ")
output_file_name = input("Enter the output CSV file name (e.g: Airbnb.csv): ")

google_play_df = scrape_google_playstore_reviews(package_name)
print(google_play_df.shape)

export_to_csv(output_file_name, google_play_df)
google_play_df.head()
```
